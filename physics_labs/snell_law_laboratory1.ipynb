```python
import pandas as pd
import math
import matplotlib.pyplot as plt
import array
import sklearn
import seaborn as sns
import numpy as np
import matplotlib
```

## יחס בין זווית פגיעה וזווית שבירה ניסוי עם צד השטוח של העדשה ##

```python
# coordinates X 
x = [0,10,20,30,40,50,60,70,80]
# coordinates Y 
y = [0,6,13,19.5,26.5,31,35.5,39.5,42]

plt.figure(figsize=(10, 4))
plt.title("יחס בין זווית פגיעה לזוות שבירה. נוסוי עם צד השטוח של העדשה"[::-1])
plt.xlabel("זווית פגיעה"[::-1])
plt.ylabel("זווית שבירה"[::-1])

plt.scatter(x, y, s=300, alpha=1,  label="יחס בין זווית פגיעה לזווית שבירה"[::-1], lw=3,  marker = '*' )
for i in range(len(x)):
    plt.scatter(x[i],y[i])
plt.legend()
plt.grid(True)
plt.show()
```

## סינוסים זוויות פגיעה ושבירה מצד השטוח ##

```python
# ערכי סינוסים של זוויות מצד שטוח
x_values = np.sin(np.array(( 0., 10., 20., 30., 40., 50., 60., 70., 80. )) * np.pi / 180.)
y_values= np.sin(np.array(( 0., 6., 13., 19.5, 26.5, 31., 35.5, 39.5, 42. )) * np.pi / 180.)

# חישוב שיפוע
slope = (y_values[-1] - y_values[0]) / (x_values[-1] - x_values[0])

# בניית גרף
x_plot = np.linspace(0, 1, 10)
y_plot = slope * (x_plot - x_values[0]) + y_values[0]


#יש לזכור את השגיאה המינימלית בחישובים בעת שימוש במכשירי מדידה, ניקח את הערך הממוצע של שגיאת הסרגל במהלך הניסוי = 0.03
yerr = [0.03,0.03,0.03,0.03,0.03,0.03,0.03,0.03,0.03]

# בניית גרף עם השגיאות
plt.scatter(x_values, y_values)
plt.errorbar(x_values, y_values, yerr=yerr, fmt='o', label='שגיאות'[::-1], color = "c")
plt.plot(x_plot, y_plot, label=f"y = {slope:.2f}x + {y_values[0]}", color = "b")
plt.xlabel("סינוס זווית פגיעה"[::-1])
plt.ylabel("סינוס זווית שבירה"[::-1])
plt.title("סינוסים זוויות פגיעה ושבירה מצד השטוח"[::-1])
plt.grid(True)
plt.legend()
plt.show()
```

## סינוסים זוויות פגיעה וזוויות שבירה בניסוי עם צד המעוגל של העדשה ##

```python
#ערכי סינוסים של זוויות מצד המעוגל
x_values = np.sin(np.array(( 0., 10., 20., 30., 40. )) * np.pi / 180.)
y_values= np.sin(np.array(( 0., 15., 30., 47.5, 72 )) * np.pi / 180.)

# שיפוע
slope = (y_values[-1] - y_values[0]) / (x_values[-1] - x_values[0])

# בניית הגרף
x_plot = np.linspace(0, 0.8, 10)
y_plot = slope * (x_plot - x_values[0]) + y_values[0]


#יש לזכור את השגיאה המינימלית בחישובים בעת שימוש במכשירי מדידה, ניקח את הערך הממוצע של שגיאת הסרגל במהלך הניסוי = 0.03
yerr = 0.03

#בניית הגרף כולל שגיאות 
plt.scatter(x_values, y_values)
plt.errorbar(x_values, y_values, yerr=yerr, fmt='o', label='שגיאות'[::-1], color = "c")
plt.plot(x_plot, y_plot, label=f"y = {slope:.2f}x + {y_values[0]}", color = "m")
plt.xlabel("סינוס זווית פגיעה"[::-1])
plt.ylabel("סינוס זווית שבירה"[::-1])
plt.title("סינוסים זוויות פגיעה וזוויות שבירה בניסוי עם צד המעוגל"[::-1])
plt.grid(True)
plt.legend()
plt.show()
```



# מקדם שבירה ניסוי עם פרספקס עברת לזר מתווך דליל לתווך צפוף #

```python
#זווית פגיעה
x_values= np.sin(np.array(( 0., 10., 20., 30., 40., 50., 60., 70., 80. )) * np.pi / 180.)
#זווית שבירה
y_values= np.sin(np.array(( 0., 6., 13., 19.5, 26.5, 31., 35.5, 39.5, 42. )) * np.pi / 180.)

# מקדם שבירה ניסוי עם פרספקס עברת לזר מתווך דליל לתווך צפוף

mekadem_shvirra = (x_values[-1] - x_values[0]) / (y_values[-1] - y_values[0])

title_2 ="מקדם שבירה של פרספקס בניסוי עם צד השטוח של פרספקס"[::-1]
print ( title_2 [::-1] , ":", round(mekadem_shvirra, 2 ))


#שיפוע של הגרף
shipua = 1/ mekadem_shvirra
shipuaa = round(shipua, 2)
title_1 = "שיפוע של הגרף"[::-1]
print( title_1[::-1], shipuaa )


#כדי לחשב את השגיאה בתוצאות המחקר, נשווה את המקדם השבירה המתקבל של הפרספקס עם ערכו הקבוע(1.4814)

title_3 = "שגיאת המקדם"[::-1]
shgiyat_amekadem = ((1.4914 - mekadem_shvirra)/ 1.4914)*100
print( "%", shgiyat_amekadem, title_3[::-1])

```



# מקדם שבירה ניסוי עם פרספקס עברת לזר מתווך צפוף לתווך דליל #

```python
import math

# זווית פגיעה
x_vv = np.sin(np.array(( 0., 10., 20., 30., 40. )) * np.pi / 180.)
#זוות שבירה
y_vv= np.sin(np.array(( 0., 15., 30., 47.5, 72 )) * np.pi / 180.)

# שיפוע של הגרף

tatmekadem = (y_vv[-1] - y_vv[0]) / (x_vv[-1] - x_vv[0])
mekadem_fat_side = round(tatmekadem, 3)
title_11 ="מקדם שבירה של פרספקס בניסוי עם צד מעוגל של פרספקס"[::-1]
print( title_11 [::-1] , ":", mekadem_fat_side)

#כדי לחשב את השגיאה בתוצאות המחקר, נשווה את המקדם השבירה המתקבל של הפרספקס עם ערכו הקבוע(1.4814)

#חישוב שגיאות

shgiyat_amekadem_fatside = (abs(1.4914 - mekadem_fat_side)/ 1.4914)*100
print("%",shgiyat_amekadem_fatside, "שגיאת המקדם")
```

## חישוב זווית הקריטית על סמך תוצאות הניסוי עם פרספקס בעברת לזר מתווך צפוף לתווך דליל ##

```python
#חישוב זווית הקריטית
sin_c = 1 / mekadem_fat_side

title_x = "זווית הקריטית על פי הנתונים של הגרף"[::-1]

# find in radians
critit_radians = math.asin(sin_c)

# find in degrees
critit_degrees = math.degrees(critit_radians)

print(title_x[::-1], critit_degrees)
title_y = "סינוס זווית קריטית ברדיאנים"[::-1]
print(sin_c, title_y[::-1])

#ערך זווית הקריטית שקייבלנו במהלך המעבדה הוא 44
#עררך הקבוע של הזווית הקריטית של פרספקס הוא 42.155
# 1/1.49 = 42.155 מעלות
shgiyat_sin_c = (abs(42.155 - 44)/ 42.155)*100
print("%",shgiyat_sin_c, "שגיאת המקדם")
```
