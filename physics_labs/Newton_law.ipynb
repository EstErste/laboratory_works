```python
def my_function(x, y):
    return ((1/((497.5+x)+y))*(y*9.8))

# Инициализируем пустой список для хранения результатов
results = []
x_values = []

# Получаем ввод от пользователя 5 раз
for _ in range(5):
    x = float(input("Введите первое число: "))
    y = float(input("Введите второе число: "))
    result = my_function(x, y)
    results.append(result)
    x_values.append(x)
print("ускорения:", results)
results_array = np.array(results)
x_values_array = np.array(x_values)
print("дополнительный вес в тележке:", x_values_array)
```

```python
from tabulate import tabulate

# Функция для вычисления значений
def calculate_values():
    x = 3.2
    increment = 0.04
    values = []

    for i in range(14):  
        values.append(x)
        x += increment

    return values

# S(y)
eteky = [ 1.85, 2.5, 3.2, 3.95, 4.8, 5.8, 6.85, 8, 9.2,10.48, 11.85, 13.35, 14.9, 16.57, 18.42]

# Комбинируем результаты функции и другие значения
combined_values = list(zip(calculate_values(), eteky))

# Заголовки для таблицы
headers = ["t", "S(y)"]

# Выводим таблицу
print(tabulate(combined_values, headers=headers, tablefmt="grid"))
timelist = calculate_values()
print(timelist)
```

```python
def calculate_speed(distance_measurements):
    # Инициализируем пустой список для измерений скорости
    speed_measurements = []

    # Итерируемся по измерениям расстояния
    for i in range(1, len(distance_measurements)):
        # Вычисляем разницу между последовательными измерениями расстояния
        distance_diff = distance_measurements[i] - distance_measurements[i - 1]
        
        # Вычисляем скорость, разделив разницу на 0.04
        speed = distance_diff / 0.04
        
        # Добавляем скорость в список
        speed_measurements.append(speed)

    return speed_measurements

# Заданные измерения расстояния
distance_measurements = [1.85, 2.5, 3.2, 3.95, 4.8, 5.8, 6.85, 8, 9.2, 10.48, 11.85, 13.35, 14.9, 16.57, 18.42]

# Вычисляем измерения скорости
speed_results = calculate_speed(distance_measurements)

# Выводим измерения скорости
print(speed_results)


# Комбинируем результаты функции и другие значения
combinedlyat = list(zip(calculate_values(), speed_results))

# Заголовки для таблицы
headers = ["t", "V"]

# Выводим таблицу
print(tabulate(combinedlyat, headers=headers, tablefmt="grid"))
```
